```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("LocalPractice") \
    .master("local[*]") \
    .getOrCreate()
sc = spark.sparkContext
print(spark.sparkContext.master)
print(spark.version)
```

```python
data = sc.textFile('/home/jovyan/data/movies.csv')
```

```python
data.map(lambda s: s.split(',')[0]) \
    .flatMap(lambda name: name.lower().split(' ')) \
    .map(lambda x: (x, 1)) \
    .reduceByKey(lambda a,b: a + b) \
    .takeOrdered(10, key=lambda x: -x[1])
```
